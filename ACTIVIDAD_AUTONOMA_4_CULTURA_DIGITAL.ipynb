```python
import time
import math
import matplotlib.pyplot as plt
```

```python
inicio = time.time()

primos = []

for num in range(1, 100000):
    es_primo = True

    if num < 2:
        es_primo = False
    else:
        for i in range(2, num):
            if num % i == 0:
                es_primo = False
                break
    
    if es_primo:
        primos.append(num)

fin = time.time()

tiempo_original = fin - inicio

print("Tiempo original:", tiempo_original)
```

```python
inicio = time.time()

primos = []

for num in range(1, 100000):
    es_primo = True

    if num < 2:
        es_primo = False
    else:
        for i in range(2, int(math.sqrt(num)) + 1):
            if num % i == 0:
                es_primo = False
                break
    
    if es_primo:
        primos.append(num)

fin = time.time()

tiempo_optimizado = fin - inicio

print("Tiempo optimizado:", tiempo_optimizado)
```

```python
print("Tiempo original:", tiempo_original)
print("Tiempo optimizado:", tiempo_optimizado)
```

```python
codigos = ["Original", "Optimizado"]
tiempos = [tiempo_original, tiempo_optimizado]

plt.bar(codigos, tiempos)

plt.title("Comparacion de tiempos")
plt.xlabel("Codigo")
plt.ylabel("Segundos")

plt.show()
```
